```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('mtcars.csv')
```

```python
data
```

```python
data.shape
```

```python
data.head()
```

```python
pd.crosstab(data['gear'],data['cyl'])
```

```python
pd.crosstab(data['gear'],data['cyl']).plot.bar() 
```

```python
data['gear'].value_counts()
```

```python
# this command gives us the pie chart for visualisation of gear data
data.gear.value_counts().plot.pie()
```

```python
plt.scatter(data.mpg,data.qsec,color='green')
```

```python
# Histogram
#google - plt.hist in matplotlib.pyplot
plt.hist(data.mpg,edgecolor='red')
```

```python
plt.hist(data['mpg'],facecolor='orange',edgecolor='white',bins=6)# bins means bars
```

```python
# Boxplot  it show the information of the outliers
plt.boxplot(data.mpg)
plt.show()
```

```python
# Boxplot  it show the information of the outliers
data.boxplot()
plt.show()
```

```python
plt.boxplot(data.mpg,vert=False,notch=True) #notch- it just visualisation like at median it goes narrow
plt.show()
```
