```python
# plot graph of no.(-100,100) and its squre, use the range function
x=list(range(-100,100))
y=[i**2 for i in x]# list comprehension

plt.plot(x,y)
```

```python
# this the exponansial gram
from math import exp
x=list(range(0,30))
y=[exp(i) for i in x]
plt.plot(x,y)
```

```python
from math import log
x=list(range(1,101)) # we have the range start from 1 not from 0
y=[log(i) for i in x]
plt.plot(x,y)
```

```python
# plot multiple graphs at a time eg-12 graphs at a time -3X4 graphs
from math import *
plt.figure(figsize=(16,9))
x=list(range(-10,11))
y=[i**2 for i in x]
plt.subplot(3,3,1)# 1 position of graph

plt.plot(x,y)

y=[i**2 for i in x]
plt.subplot(3,4,7) # 7 positions of graphs
plt.plot(x,y)

y=[sin(i) for i in x]
plt.subplot(3,4,11)# 10 positions of graph
plt.plot(x,y)
```
